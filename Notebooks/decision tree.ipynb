```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets\\asiacup.csv')
df.head()
```

```python
df = df.loc[ df['Team']=='India']
df.head()
```

```python
features = ['Avg Bat Strike Rate','Wicket Taken','Run Scored','Fours','Sixes','Result']
df = df.dropna()
df = df[features].copy()
df.head()
```

```python
df['Result'] = df['Result'].map({'Win':1,'Lose':0, 'No Result':-1})
df.head()
```

```python
df = df.loc[df.Result!=-1]
x = df.loc[ :,'Avg Bat Strike Rate':'Sixes'].values
y = df.loc[ :, 'Result':].values
# df[df.isna().any(axis=1)]
df = df.dropna()
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(x,y,test_size = 0.25, random_state= 0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier = classifier.fit(X_train,y_train)
```

```python
#prediction
y_pred = classifier.predict(X_test)#Accuracy
from sklearn import metrics
print('Accuracy Score:', metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
