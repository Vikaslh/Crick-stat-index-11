```python
import pandas as pd
```

```python
#import csv files
batsmen = pd.read_csv('batsmanT20.csv')
bowlers = pd.read_csv('BowlerT20.csv')
bowlers
```

```python
# Creating a dataframe for batsmen and bowlers with specified columns.
batdf = pd.DataFrame( batsmen, columns=['Player Name','Batting Average','Balls Faced', 'Fours', 'Sixes'])
balldf = pd.DataFrame( bowlers, columns=['Player Name','Economy Rate','Wickets','Bowling Average'])
```

```python
batdf
```

```python
player = pd.concat([batdf,balldf], ignore_index = True, sort = False)
player
```
