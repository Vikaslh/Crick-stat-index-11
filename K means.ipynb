```python
import pandas as pd
batsman = pd.read_csv('batsmanT20.csv')
batsman.head()
```

```python
features = ['Batting Average','Strike Rate','Runs', 'Not Outs']
batsman = batsman.dropna(subset=features)
batdf = batsman[features].copy()
batdf = batdf.set_index(batsman['Player Name'])
# batdf.index.names = [None]
batdf
```

```python
batdf = batdf.drop('Mosaddek Hossain ')
batdf
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 10), copy=True)
scaled = pd.DataFrame( scaler.fit_transform(batdf), columns=batdf.columns, index=batdf.index)
scaled
```

```python
def random_centroids(data,k):
    centroids = []
    for i in range(k):
#         centroids.append( scaled.apply(lambda x:  float(x.sample())) )
        centroids.append(scaled.apply(lambda x:  float(x.sample().iloc[0])))
    return pd.concat( centroids, axis = 1)
```

```python
centroids = random_centroids(scaled,3)
centroids
```
