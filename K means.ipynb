```python
import pandas as pd # importing pandas to work with dataframe
batsman = pd.read_csv('batsmanT20.csv')
batsman.head()
```

```python
features = ['Batting Average','Strike Rate','Runs', 'Not Outs']
batsman = batsman.dropna(subset=features) # remove rows not dont have numerical value from the features.
batdf = batsman[features].copy()
batdf = batdf.set_index(batsman['Player Name']) # set players name for corresponding rows.
# batdf.index.names = [None]
batdf
```

```python
batdf = batdf.drop('Mosaddek Hossain ')
batdf
```

```python
# scaling the data ...
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 10), copy=True)
scaled = pd.DataFrame( scaler.fit_transform(batdf), columns=batdf.columns, index=batdf.index)
scaled
```

```python
# visualization
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
plt.scatter(scaled['Runs'],scaled['Strike Rate'],scaled['Batting Average'],scaled['Not Outs'])
```

```python
# clustering unscaled dataset
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
grp = km.fit_predict(batdf[features])
grp
```

```python
batdf['Group'] = grp
batdf
```

```python
# clustering unscaled dataset.
scaled['Group'] = km.fit_predict(scaled[features])
scaled
```

```python
grp_0 = scaled.loc[scaled['Group']==0]
grp_1 = scaled.loc[scaled['Group']==1]
grp_2 = scaled.loc[scaled['Group']==2]
```
